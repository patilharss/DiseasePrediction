```python
import pandas as pd
from tensorflow import keras
```

```python
df=pd.read_csv('diabetes.csv')
```

```python
df.head()
```

```python
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
X = df.to_numpy()[:,0:8] 
Y = df.to_numpy()[:,8]
seed = 42
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = seed)
print (f'Shape of Train Data : {X_train.shape}')
print (f'Shape of Test Data : {X_test.shape}')
```

```python
type(df['BMI'][0])
```

```python
model = keras.Sequential([
    keras.layers.Dense(12, input_dim = (8), activation = 'relu'),
    keras.layers.Dense(8, activation = 'relu'),
    keras.layers.Dense(1, activation = 'sigmoid'),
])
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=1000, batch_size=32, verbose = 1)
```

```python
scores = model.evaluate(X_test, y_test)
print (f'{model.metrics_names[1]} : {round(scores[1]*100, 2)} %')
```

```python
model.predict([[1.8,85.2,66.3,29.2,0,26.6,0.351,31]])
```

```python
import tensorflow as tf
from tensorflow import lite

kf='diabetes.h5'
tf.keras.models.save_model(model,kf)
converter=lite.TFLiteConverter.from_keras_model(model)
tfmodel=converter.convert()
open('diabetes.tflite','wb').write(tfmodel)
```
