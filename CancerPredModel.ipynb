```python
df=pd.read_csv('cancer.csv')
```

```python
d=pd.get_dummies(df.diagnosis)
```

```python

df=pd.concat([df,d],1)
```

```python
#df.drop(['diagnosis','B'],axis=1,inplace=True)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X = df.to_numpy()[:,0:30] 
Y = df.to_numpy()[:,30]
seed = 42
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = seed)
print (f'Shape of Train Data : {X_train.shape}')
print (f'Shape of Test Data : {X_test.shape}')
```

```python
model = keras.Sequential([
    keras.layers.Dense(24, input_dim = (30), activation = 'relu'),
    keras.layers.Dense(12, activation = 'relu'),
    keras.layers.Dense(1, activation = 'sigmoid'),
])
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=350)
```

```python
scores = model.evaluate(X_test, y_test)
print (f'{model.metrics_names[1]} : {round(scores[1]*100, 2)} %')
```

```python
kf='cancer.h5'
tf.keras.models.save_model(model,kf)
converter=lite.TFLiteConverter.from_keras_model(model)
tfmodel=converter.convert()
open('cancer.tflite','wb').write(tfmodel)
```
